```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the license plate image
img = cv2.imread('UKLicencePlateDataset/yellowplate_normal/AA04QZH.png')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,cmap='gray' )
```

```python
#Maximize Contrast

structuringElement = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

imgTopHat = cv2.morphologyEx(gray, cv2.MORPH_TOPHAT, structuringElement)
imgBlackHat = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, structuringElement)

imgGrayscalePlusTopHat = cv2.add(gray, imgTopHat)
gray = cv2.subtract(imgGrayscalePlusTopHat, imgBlackHat)
```

```python

plt.figure(figsize=(12, 10))
plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.savefig('Car-Contrast.png',bbox_inches = 'tight')
plt.show()
```

```python
#Adaptive Thresholding

img_blurred = cv2.GaussianBlur(gray, ksize=(5, 5), sigmaX=0)

img_thresh = cv2.adaptiveThreshold(
    img_blurred, 
    maxValue=255.0, 
    adaptiveMethod=cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
    thresholdType=cv2.THRESH_BINARY_INV, 
    blockSize=19, 
    C=9
)

plt.figure(figsize=(12, 10))
plt.imshow(img_thresh, cmap='gray')
plt.axis('off')
#plt.savefig('Car-Adaptive-Thresholding.png',bbox_inches = 'tight')
plt.show()
```

```python
height, width, channel = img.shape
```

```python
#Finding Contours to locate plate

contours, _= cv2.findContours(
    img_thresh, 
    mode=cv2.RETR_LIST, 
    method=cv2.CHAIN_APPROX_SIMPLE
)

temp_result = np.zeros((height, width, channel), dtype=np.uint8)

cv2.drawContours(temp_result, contours=contours, contourIdx=-1, color=(255, 255, 255))

plt.figure(figsize=(12, 10))
plt.imshow(temp_result)
plt.axis('off')
#plt.savefig('Car-Contours.png',bbox_inches = 'tight')
plt.show()
```

```python
#Data Preparation

temp_result = np.zeros((height, width, channel), dtype=np.uint8)

contours_dict = []

for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(temp_result, pt1=(x, y), pt2=(x+w, y+h), color=(255, 255, 255), thickness=2)
    
    # insert to dict
    contours_dict.append({
        'contour': contour,
        'x': x,
        'y': y,
        'w': w,
        'h': h,
        'cx': x + (w / 2),
        'cy': y + (h / 2)
    })

plt.figure(figsize=(12, 10))
plt.imshow(temp_result, cmap='gray')
plt.axis('off')
#plt.savefig('Car-Boxes.png',bbox_inches = 'tight')
plt.show()
```

```python
#Selecting Boxes by Char Size

MIN_AREA = 80
MIN_WIDTH, MIN_HEIGHT = 2, 8
MIN_RATIO, MAX_RATIO = 0.25, 1.0

possible_contours = []

cnt = 0
for d in contours_dict:
    area = d['w'] * d['h']
    ratio = d['w'] / d['h']
    
    if area > MIN_AREA \
    and d['w'] > MIN_WIDTH and d['h'] > MIN_HEIGHT \
    and MIN_RATIO < ratio < MAX_RATIO:
        d['idx'] = cnt
        cnt += 1
        possible_contours.append(d)
        
```

```python
# visualize possible contours
temp_result = np.zeros((height, width, channel), dtype=np.uint8)

for d in possible_contours:
#    cv2.drawContours(temp_result, d['contour'], -1, (255, 255, 255))
    cv2.rectangle(temp_result, pt1=(d['x'], d['y']), pt2=(d['x']+d['w'], d['y']+d['h']), color=(255, 255, 255), thickness=2)

plt.figure(figsize=(12, 10))
plt.imshow(temp_result, cmap='gray')
plt.axis('off')
plt.savefig('Car-Boxes-byCharSize.png',bbox_inches = 'tight')
plt.show()
```

```python
#Selecting Boxes by Arrangement of Contours

MAX_DIAG_MULTIPLYER = 5 # 5
MAX_ANGLE_DIFF = 12.0 # 12.0
MAX_AREA_DIFF = 0.5 # 0.5
MAX_WIDTH_DIFF = 0.8
MAX_HEIGHT_DIFF = 0.2
MIN_N_MATCHED = 3 # 3

def find_chars(contour_list):
    matched_result_idx = []
    
    for d1 in contour_list:
        matched_contours_idx = []
        for d2 in contour_list:
            if d1['idx'] == d2['idx']:
                continue

            dx = abs(d1['cx'] - d2['cx'])
            dy = abs(d1['cy'] - d2['cy'])

            diagonal_length1 = np.sqrt(d1['w'] ** 2 + d1['h'] ** 2)

            distance = np.linalg.norm(np.array([d1['cx'], d1['cy']]) - np.array([d2['cx'], d2['cy']]))
            if dx == 0:
                angle_diff = 90
            else:
                angle_diff = np.degrees(np.arctan(dy / dx))
            area_diff = abs(d1['w'] * d1['h'] - d2['w'] * d2['h']) / (d1['w'] * d1['h'])
            width_diff = abs(d1['w'] - d2['w']) / d1['w']
            height_diff = abs(d1['h'] - d2['h']) / d1['h']

            if distance < diagonal_length1 * MAX_DIAG_MULTIPLYER \
            and angle_diff < MAX_ANGLE_DIFF and area_diff < MAX_AREA_DIFF \
            and width_diff < MAX_WIDTH_DIFF and height_diff < MAX_HEIGHT_DIFF:
                matched_contours_idx.append(d2['idx'])

        # append this contour
        matched_contours_idx.append(d1['idx'])

        if len(matched_contours_idx) < MIN_N_MATCHED:
            continue

        matched_result_idx.append(matched_contours_idx)

        unmatched_contour_idx = []
        for d4 in contour_list:
            if d4['idx'] not in matched_contours_idx:
                unmatched_contour_idx.append(d4['idx'])

        unmatched_contour = np.take(possible_contours, unmatched_contour_idx)
        
        # recursive
        recursive_contour_list = find_chars(unmatched_contour)
        
        for idx in recursive_contour_list:
            matched_result_idx.append(idx)

        break

    return matched_result_idx
    
result_idx = find_chars(possible_contours)

matched_result = []
for idx_list in result_idx:
    matched_result.append(np.take(possible_contours, idx_list))
```

```python
import cv2
import numpy as np

def calculate_variance_features(img):
    # Calculate the dimensions of each part
    height, width, _ = img.shape
    part_height = height // 5  # Divide the image height into 5 parts
    part_width = width // 8  # Divide the image width into 8 parts

    # Initialize the list to store variance features
    variance_features = []

    # Loop through the image parts
    for i in range(5):
        for j in range(8):
            # Calculate the coordinates of the current part
            x_start = j * part_width
            x_end = x_start + part_width
            y_start = i * part_height
            y_end = y_start + part_height

            # Extract the part from the image
            part = img[y_start:y_end, x_start:x_end]

            # Calculate the variance of the part
            gray_part = cv2.cvtColor(part, cv2.COLOR_BGR2GRAY)
            variance = np.var(gray_part)

            # Add the variance feature to the list
            variance_features.append(variance)

    return variance_features
```

```python
import os
import csv
import string

# Specify the directory path where your images are stored
image_directory = 'UKLicencePlateDataset/whiteplate_normal'

# Get a list of image filenames in the directory
image_filenames = os.listdir(image_directory)

# Create a CSV file and write the header row
csv_filename = 'features3.csv'
feature_names = [f'Feature{i+1}' for i in range(81)]  # Replace 'Feature' with your desired feature name prefix
header_row = feature_names + ['Label']

# Define the set of valid characters
valid_characters = set(string.ascii_lowercase + string.digits)  # 'abcdefghijklmnopqrstuvwxyz0123456789'

# Track the encountered characters
encountered_characters = set()

with open(csv_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header_row)

    # Loop through the images
    for image_filename in image_filenames:
        # Assuming the image filename corresponds to the character labels
        char_img_name = os.path.splitext(image_filename)[0]  # Get the image name without extension

        # Load the image
        img = cv2.imread(os.path.join(image_directory, image_filename))

        # Perform character detection and get matched_result

        # Extract individual characters and preprocess
        hog_features = []
        labels = []

        for r in matched_result:
            for d in r:
                x, y, w, h = d['x'], d['y'], d['w'], d['h']
                char_img = img[y:y+h, x:x+w]

                # Preprocess the character image
                feature = preprocess(char_img)
                hog_features.append(feature)

                # Extract the label based on the order of characters in the image name
                label = char_img_name[len(labels)]
                if label.lower() in valid_characters and label not in encountered_characters:
                    labels.append(label)
                    encountered_characters.add(label)
                lenn = len(feature)
                #print(f'Label={label} and {lenn} features')
                #print(f'Features={feature}')

        # Write the features and labels to the CSV file
        with open(csv_filename, mode='a', newline='') as file:
            writer = csv.writer(file)
            for i in range(len(hog_features)):
                writer.writerow(list(hog_features[i]) + [labels[i]])

        # Stop the loop if all alphabets and numbers have been encountered
        if len(encountered_characters) == len(valid_characters):
            break

        #print(f"Features and labels for image '{char_img_name}' added to the CSV file.")

print(f"CSV file '{csv_filename}' created successfully.")
```

```python
import csv
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the CSV data
csv_filename = 'features6.csv'
data = []
labels = []

with open(csv_filename, mode='r') as file:
    reader = csv.reader(file)
    header_row = next(reader)  # Skip the header row

    for row in reader:
        features = np.array(row[:-1], dtype=np.float32)  # Convert features to numpy array
        label = row[-1]
        data.append(features)
        labels.append(label)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

# Train the SVM classifier
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)

# Make predictions on the test set
y_pred = svm.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Predict some images from your path
#path = 'your_image_path'  # Replace with the actual path to your images

# Load and preprocess the images
# ...

# Extract features from the preprocessed images
# ...

# Make predictions using the trained SVM classifier
#predictions = svm.predict(features)

# Print the predicted labels
#print("Predictions:")
#for image_path, label in zip(image_paths, predictions):
 #   print(f"{image_path}: {label}")
```

```python
import os
import csv
import cv2
import numpy as np
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input

# Specify the directory path where your images are stored
image_directory = 'UKLicencePlateDataset/whiteplate_normal'

# Get a list of image filenames in the directory
image_filenames = os.listdir(image_directory)

# Load the VGG16 model
vgg_model = VGG16(weights='imagenet', include_top=False)

# Create a CSV file and write the header row
csv_filename = 'features_vgg.csv'
header_row = ['Label'] + [f'Feature{i+1}' for i in range(7*7*512)]  # Update the feature count based on VGG16's output shape

with open(csv_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header_row)

    # Loop through the images
    for image_filename in image_filenames:
        # Assuming the image filename corresponds to the character labels
        char_img_name = os.path.splitext(image_filename)[0]  # Get the image name without extension

        # Load the image
        img = cv2.imread(os.path.join(image_directory, image_filename))

        # Determine the input shape based on the image shape
        input_shape = img.shape

        # Resize the image to match the input shape of VGG16
        img = cv2.resize(img, (input_shape[1], input_shape[0]))

        # Preprocess the image
        img = preprocess_input(img)

        # Expand the dimensions to match the input shape of VGG16
        img = np.expand_dims(img, axis=0)

        # Extract features using VGG16
        features = vgg_model.predict(img)

        # Flatten the features
        features = features.flatten()

        # Write the label and features to the CSV file
        writer.writerow([char_img_name] + list(features))

    print(f"CSV file '{csv_filename}' created successfully.")
```
